```python
import tensorflow as tf

from tensorflow import keras

import numpy as np

from sklearn.preprocessing import StandardScaler


# Datos de ejemplo (caractersticas de las flores)

caracteristicas = np.array([[1.2, 0.7], [0.5, 0.2], [0.9, 0.8], [1.5, 1.0], [1.0, 0.5]])

# Etiquetas de clase (0 para "flores rojas" y 1 para "flores azules")

etiquetas = np.array([0, 1, 0, 1, 0])


# Escalar las caractersticas

scaler = StandardScaler()

caracteristicas_scaled = scaler.fit_transform(caracteristicas)


# Construir el modelo

model = keras.Sequential([

    keras.layers.Dense(units=2, activation='relu', input_shape=(2,)),  # Capa oculta con 2 neuronas y funcin de activacin ReLU

    keras.layers.Dense(units=1, activation='sigmoid')  # Capa de salida con 1 neurona y funcin de activacin sigmoide para clasificacin binaria

])


# Compilar el modelo

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


# Entrenar el modelo

model.fit(caracteristicas_scaled, etiquetas, epochs=1000)


# Escalar las caractersticas de una nueva flor (ejemplo)

nueva_flor = np.array([[1.3, 0.6]])

nueva_flor_scaled = scaler.transform(nueva_flor)


# Hacer una prediccin para la nueva flor escalada

prediccion = model.predict(nueva_flor_scaled)


# Imprimir la prediccin

print("Prediccin:", prediccion)
```

```python
import tensorflow as tf

from tensorflow import keras

import numpy as np

from sklearn.preprocessing import StandardScaler


# Datos de ejemplo (caractersticas de las flores)

caracteristicas = np.array([[1.2, 0.7], [0.5, 0.2], [0.9, 0.8], [1.5, 1.0], [1.0, 0.5]])

# Etiquetas de clase (0 para "flores rojas" y 1 para "flores azules")

etiquetas = np.array([0, 1, 0, 1, 0])


# Escalar las caractersticas

scaler = StandardScaler()

caracteristicas_scaled = scaler.fit_transform(caracteristicas)


# Construir el modelo

model = keras.Sequential([

    keras.layers.Dense(units=2, activation='relu', input_shape=(2,)),  # Capa oculta con 2 neuronas y funcin de activacin ReLU

    keras.layers.Dense(units=1, activation='sigmoid')  # Capa de salida con 1 neurona y funcin de activacin sigmoide para clasificacin binaria

])


# Compilar el modelo

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


# Entrenar el modelo

model.fit(caracteristicas_scaled, etiquetas, epochs=1000)


# Escalar las caractersticas de una nueva flor (ejemplo)

nueva_flor = np.array([[1.3, 0.6]])

nueva_flor_scaled = scaler.transform(nueva_flor)


# Hacer una prediccin para la nueva flor escalada

prediccion_probabilidad = model.predict(nueva_flor_scaled)


# Aplicar un umbral de decisin (0.5) para obtener una etiqueta de clase

umbral = 0.5

prediccion_etiqueta = 1 if prediccion_probabilidad >= umbral else 0


# Imprimir la etiqueta de clase predicha

print("Etiqueta de clase predicha:", prediccion_etiqueta)
```
