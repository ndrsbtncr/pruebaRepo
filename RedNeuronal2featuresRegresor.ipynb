```
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras

# Datos de ejemplo
areas = np.array([120, 150, 180, 200, 220, 250, 280, 300, 320, 350, 380, 400, 420, 450, 480, 500, 520, 550, 580, 600])
precios = np.array([250, 320, 370, 400, 430, 480, 520, 550, 580, 620, 650, 680, 700, 730, 760, 780, 800, 830, 860, 880])

# Escalar los datos
scaler_x = StandardScaler()
scaler_y = StandardScaler()

areas_scaled = scaler_x.fit_transform(areas.reshape(-1, 1))
precios_scaled = scaler_y.fit_transform(precios.reshape(-1, 1))

# Construir el modelo
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])  # Una capa densa con una neurona])

# Compilar el modelo
model.compile(optimizer='sgd', loss='mean_squared_error')  # Usando el descenso de gradiente estocástico y error cuadrático medio

# Entrenar el modelo
model.fit(areas_scaled, precios_scaled, epochs=1000)  # Entrenar durante 1000 épocas

# Predicción para una nueva casa con área de 650 metros cuadrados
area_nueva_scaled = scaler_x.transform(np.array([650]).reshape(-1, 1))
precio_predicho_scaled = model.predict(area_nueva_scaled)

# Desescalar la predicción
precio_predicho = scaler_y.inverse_transform(precio_predicho_scaled)

print("El precio estimado para una casa con un área de 650 metros cuadrados es:", precio_predicho[0][0], "miles de dólares")
```

```
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras

# Datos de ejemplo con dos características (áreas y número de habitaciones)
datos_de_entrada = np.array([[120, 3],
                            [150, 4],
                            [180, 3],
                            [200, 4],
                            [220, 3],
                            [250, 4],
                            [280, 4],
                            [300, 5],
                            [320, 5],
                            [350, 4],
                            [380, 5],
                            [400, 4],
                            [420, 5],
                            [450, 5],
                            [480, 6],
                            [500, 6],
                            [520, 5],
                            [550, 6],
                            [580, 6],
                            [600, 7]])

precios = np.array([250, 320, 370, 400, 430, 480, 520, 550, 580, 620, 650, 680, 700, 730, 760, 780, 800, 830, 860, 880])

# Escalar los datos
scaler_x = StandardScaler()
scaler_y = StandardScaler()

datos_escalados = scaler_x.fit_transform(datos_de_entrada)
precios_scaled = scaler_y.fit_transform(precios.reshape(-1, 1))

# Construir el modelo
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=(2,))  # Dos características de entrada
])

# Compilar el modelo
model.compile(optimizer='sgd', loss='mean_squared_error')

# Entrenar el modelo
model.fit(datos_escalados, precios_scaled, epochs=1000)

# Predicción para una nueva casa con área de 650 metros cuadrados y 5 habitaciones
nueva_casa = np.array([[650, 5]])
nueva_casa_scaled = scaler_x.transform(nueva_casa)
precio_predicho_scaled = model.predict(nueva_casa_scaled)

# Desescalar la predicción
precio_predicho = scaler_y.inverse_transform(precio_predicho_scaled)

print("El precio estimado para una casa con un área de 650 metros cuadrados y 5 habitaciones es:", precio_predicho[0][0], "miles de dólares")
```
